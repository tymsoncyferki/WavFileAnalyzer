```python
from analysis import FileAnalyzer
import os
import matplotlib.pyplot as plt
```

```python
def plot_f0(analyzer):
    f0, time = analyzer.f0()

    plt.figure(figsize=(10, 4))
    plt.plot(time, f0)
    plt.xlabel("Czas [s]")
    plt.ylabel("Częstotliwość [Hz]")
    plt.title("Częstotliwość krtaniowa f0 na podstawie cepstrum")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

def plot_waveform_freq(analyzer):
    energy, time = analyzer.waveform_freq()
    
    plt.figure(figsize=(10, 4))
    plt.plot(time, energy)
    plt.xlabel("Częstotliwość [Hz]")
    plt.ylabel("Amplituda")
    plt.title("Amplituda sygnału na poziomie klipu")
    plt.xscale('log')
    plt.xlim([1,10000])
    plt.legend()
    plt.grid()
    plt.show()

def plot_spectrogram(analyzer, window_type):
    spec_db, time_axis, freq_axis = analyzer.spectrogram(window_type)

    fig, ax = plt.subplots()

    ax.clear()

    img = ax.imshow(spec_db, origin='lower', aspect='auto',
                    extent=[time_axis[0], time_axis[-1], freq_axis[0], freq_axis[-1]])
    ax.set_xlabel("Czas [s]")
    ax.set_ylabel("Częstotliwość [Hz]")
    ax.set_title("Spektrogram")

    plt.ylim(0, 3000)

    colorbar = fig.colorbar(img, ax=ax, label="Amplituda [dB]")

    plt.show()

def plot_fc(analyzer):
    fc, time = analyzer.fc()
    
    plt.figure(figsize=(10, 4))
    plt.plot(time, fc)
    plt.xlabel("Czas [s]")
    plt.ylabel("Hz")
    plt.title("Frequency centroid na poziomie ramek")
    plt.legend()
    plt.grid()
    plt.show()

def plot_bandwidth(analyzer):
    bw, time = analyzer.bw()
    
    plt.figure(figsize=(10, 4))
    plt.plot(time, bw)
    plt.xlabel("Czas [s]")
    plt.ylabel("Hz")
    plt.title("Efektywne szerokość pasma na poziomie ramek")
    plt.legend()
    plt.grid()
    plt.show()

def plot_ersb(analyzer):
    ersb1, ersb2, ersb3, time = analyzer.ersb()
    
    plt.figure(figsize=(10, 4))
    plt.plot(time, ersb1, label="ERSB1")
    plt.plot(time, ersb2, label="ERSB2")
    plt.plot(time, ersb3, label="ERSB3")
    plt.xlabel("Czas [s]")
    plt.ylabel("Stosunek energii")
    plt.title("Stosunek energii w pasmach częstotliwości (ERSB) w czasie")
    plt.legend()
    plt.grid()
    plt.show()

def plot_sfm(analyzer):
    sfm, time = analyzer.sfm()
    
    plt.figure(figsize=(10, 4))
    plt.plot(time, sfm)
    plt.xlabel("Czas [s]")
    plt.ylabel("SFM")
    plt.title("Miara płaskości widma w czasie")
    plt.legend()
    plt.grid()
    plt.show()

def plot_sfm_bands(analyzer):
    sfmb1, sfmb2, sfmb3, time = analyzer.sfm_bands()
    
    plt.figure(figsize=(10, 4))
    plt.plot(time, sfmb1, label="SFM B1")
    plt.plot(time, sfmb2, label="SFM B2")
    plt.plot(time, sfmb3, label="SFM B3")
    plt.xlabel("Czas [s]")
    plt.ylabel("SFM")
    plt.title("Miara płaskości widma w pasmach częstotliwości (SFM subbands)")
    plt.legend()
    plt.grid()
    plt.show()

def plot_scf_bands(analyzer):
    scfb1, scfb2, scfb3, time = analyzer.scf_bands()
    
    plt.figure(figsize=(10, 4))
    plt.plot(time, scfb1, label="SCF B1")
    plt.plot(time, scfb2, label="SCF B2")
    plt.plot(time, scfb3, label="SCF B3")
    plt.xlabel("Czas [s]")
    plt.ylabel("SCF")
    plt.title("Spectral Crest Factor w pasmach częstotliwości w czasie")
    plt.legend()
    plt.grid()
    plt.show()
```
